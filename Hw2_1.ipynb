```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)"

```python
url = "https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8"

birth_rate_table = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
```

Вивести перші рядки таблиці за допомогою методу head

```python
print(birth_rate_table.head())
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape)

```python
print(birth_rate_table.shape)
```

Замініть у таблиці значення "—" на значення NaN

```python
birth_rate_table_cleared = birth_rate_table.replace("—", np.nan)
display(birth_rate_table_cleared)
```

Визначте типи всіх стовпців за допомогою dataframe.dtypes

```python
display(birth_rate_table.dtypes)
```

Замініть типи нечислових колонок на числові. Підказка - це колонки, де знаходився символ "—"

```python
nan_columns = birth_rate_table_cleared.columns[birth_rate_table_cleared.isnull().any()].tolist()
birth_rate_table_cleared[nan_columns] = birth_rate_table_cleared[nan_columns].astype(float)

display(birth_rate_table_cleared.dtypes)
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum)

```python
missing_values = birth_rate_table_cleared.isnull().sum()
print(f"Количество пропусков в каждой колонке:\n{missing_values}")
```

```python
total_values = birth_rate_table_cleared.shape[0]
missing_ratio = (missing_values / total_values) * 100

print(f"Доля пропусков в каждой колонке:\n{missing_ratio}")
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці

```python
birth_rate_table_restricted = birth_rate_table_cleared.drop(27)
display(birth_rate_table_restricted)
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna)

```python
test = birth_rate_table_cleared['2012'].mean()
print(test)
```

```python
numerical_columns = birth_rate_table_restricted.drop(columns=['Регіон'])
filled_table = numerical_columns.fillna(round(numerical_columns.mean(), 2))

birth_rate_table_restricted_filled = pd.concat(
    [birth_rate_table_restricted['Регіон'],
     filled_table],
    axis=1,
    )

display(birth_rate_table_restricted_filled)
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні

```python
avg_birth_2019 = birth_rate_table_cleared['2019'][27]
display(birth_rate_table_cleared[birth_rate_table_cleared['2019']>avg_birth_2019].loc[:, ['Регіон', '2019']])
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
birth_max_2014 = birth_rate_table_cleared['2014'].max()

display(birth_rate_table_cleared[birth_rate_table_cleared['2014']==birth_max_2014].loc[:, ['Регіон', '2014']])
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році

```python
data_2019 = birth_rate_table_cleared.loc[:26, ['Регіон', '2019']].dropna()

fig, ax = plt.subplots(figsize=(10, 6), dpi=80)

ax.bar(
    data_2019['Регіон'],
    data_2019['2019'],
    edgecolor="white",
    width=.7,
    linewidth=0.2,
    color='#6a4c93',
    )

plt.xlabel("Region", fontsize="large", color="midnightblue")
plt.ylabel("Рівень народжуваності ", fontsize="large", color="midnightblue")
plt.title("діаграмa народжуваності по регіонах у 2019 році", fontsize=15)

plt.xticks(rotation=90)

plt.show()
```

# Візуалізація

додайте від 3 до 5 графіків функцій різного типу на ваш вибір. Задайте графікам оформлення, щоб кожен графік у своїй домашній роботі чимось відрізнявся і не був схожим на інші. Можна використовувати як matplotlib, так і seaborn.

```python
data_columns = birth_rate_table_cleared.columns[1:]
reversed_columns = data_columns[::-1]
reversed_data = birth_rate_table_cleared[['Регіон'] + list(reversed_columns)]

sns.set_style("whitegrid")

plt.figure(figsize=(20, 8))

sns.heatmap(
    reversed_data.set_index('Регіон').T,
    cmap='viridis',
    annot=True,
    fmt=".2f",
    square=True,
    )

plt.xlabel('Регіон', fontsize=12)
plt.ylabel('Год', fontsize=12)
plt.title('Корреляция рождаемости между регионами и годами', fontsize=14)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    data=reversed_data.dropna(),
    orient='h',
    palette='viridis',
    )

plt.title('Розподілення рівня народжуваності по роках')
plt.xlabel('Рівень народжуваності')
plt.ylabel('Рік')
plt.grid(axis='x')
plt.show()
```

```python
reversed_data.set_index('Регіон', inplace=True)
```

```python
plt.figure(figsize=(12, 8))
for region in reversed_data.index:

    plt.plot(reversed_data.columns[1:],
             reversed_data.loc[region][1:],
             marker='o',
             label=region,
             )

plt.title('Динамика рівня народжуваності в різних регионах')
plt.xlabel('Рік')
plt.ylabel('Рівень народжуваності')
plt.xticks(rotation=45)
plt.legend(title='Регіон', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
